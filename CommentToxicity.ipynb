```python
from keras.layers import LSTM,Dropout,Bidirectional,Dense,Embedding
from keras.models import Sequential
```

The analysis of sequential data such as text sentences, time-series, and other discrete sequence data prompted the development of Sequence Models. These models are better suited to handle sequential data, whereas Convolutional Neural Networks are better suited to treat spatial data.

The crucial element to remember about sequence models is that the data we’re working with are no longer independently and identically distributed (i.i.d.) samples, and the data are reliant on one another due to their sequential order. For speech recognition, voice recognition, time series prediction, and natural language processing, sequence models are particularly popular.

Sequential model: RNN for time series data
Many-to-one

This is used to categorize emotions. A succession of words or even paragraphs of words is anticipated as input. The result can be a continuous-valued regression output that represents the likelihood of having a favourable attitude.
The name recurrent comes from the fact that it receives two inputs, one from the user and the other from the previous output (output goes as input).

LSTM:
In a cell of the LSTM neural network, the first step is to decide whether we should keep the information from the previous time step or forget it. Here is the equation for forget gate.
The input gate is used to quantify the importance of the new information carried by the input. Here is the equation of the input gate
New informations:
Nt = tanh(xt * Uc + H(t-1)*Wc)
if Nt<0 the info is subtracted from the cell state and if the value is positive, the information is added to the cell state at the current timestamp.



Hidden Layers:
Rectified Linear Unit (ReLU) is a popular choice for activation functions in hidden layers due to its simplicity and efficiency in training deep networks.

```python
model = Sequential()
# Create a embedding layer
model.add(Embedding(MAX_WORDS+1,32))
# 32 different units
# tanh - GPU acceleratiosn required by LSTM
model.add(Bidirectional(LSTM(32,activation='tanh')))
# FEATURE extractors Full connected layers
model.add(Dense(128,activation='relu'))
model.add(Dense(256,activation='relu'))
model.add(Dense(128,activation='relu'))
# Final layer
# 6 units in our output y[0] = [0,0,0,0,0,0]
model.add(Dense(6,activation='sigmoid'))
```

```python
model.compile(loss='BinaryCrossentropy',optimizer='Adam')
```
